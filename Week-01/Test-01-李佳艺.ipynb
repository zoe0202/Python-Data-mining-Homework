解法一

```python
x=False
for i in range(1,4):
    name=input("请输入用户名：")
    password=input("请输入密码：")
    if name=="LEE" and password=="12345":
        print("登陆成功！")
        x=True
        break
    else:
        if i!=3:
            print("用户名或密码错误，你还有%d次机会"%(3-i))
if not x:
    print("因连续3次输入用户名或密码错误，账号已锁死")
```

```python
i=1
while i<=3:
    name=input("请输入用户名：")
    password=input("请输入密码：")
    i=i+1
    if name=="LEE" and password=="12345":
        print("登陆成功！")
        break
    else:
        print("用户名或密码错误，若连续3次错误则自动锁死账号")
while i>3 and i<5:
    if name=="LEE" and password=="12345":
        print("登陆成功！")
        break
    else:print("因连续3次输入用户名或密码错误，账号已锁死")
    break
```
